# Data Preprocessing

```python
df.columns
```

**Changing columns names**

```python
df.columns = ["Satisfaction", "Gender", "Customer_Type", "Age", "Type_Of_Travel", "Class", "Flight_Distance", "Seat_Comfort", "Departure/Arrival_Time_Convenient", "Food_And_Drink", "Gate_Location", "Inflight_Wifi_Service", "Inflight_Entertainment", "Online_Support", "Ease_Of_Online_Booking", "On-Board_Service", "Leg_Room_Service", "Baggage_Handling", "Checkin_Service", "Cleanliness", "Online_Boarding", "Departure_Delay_In_Minutes", "Arrival_Delay_In_Minutes"]
```

```python
df.columns
```

```python
df1 = df
df1.head()
```

```python
df1.dtypes
```

It is necessary to change categorical columns to numerical values so that the model can utilize those columns, and which can help us in correlation matrix to identify more relationships.

Thos columns are `Satisfaction`, `Gender`, `Customer_Type`, `Type_Of_Travel` and `Class`

```python
from sklearn.preprocessing import LabelEncoder

encoder = LabelEncoder()
```

```python
categorical_columns = ["Satisfaction", "Gender", "Customer_Type", "Type_Of_Travel", "Class"]
for i in categorical_columns:
  print(df1[i].unique())
```

```python
for i in categorical_columns:
  df1[i] = encoder.fit_transform(df1[i])

df1.head()
```

Show the correlation matrix

```python
plt.figure(figsize=(20,20))
sns.heatmap(df1.corr(), square=True, annot=True, linewidths=0.5, cmap="seismic")
```

Results:

1.   Firstly, we can see that Departure delay in minutes and Arrival delay in minutes are highly correlated (96%) which makes sense as if a flight starts late it will most likely land late.
2.   Secondly, we can see that Seat Comfort and Food and Drink columns are highly correlated (72%).



```python
df1.head()
```

Next we see that columns Age, Flight Distance, Arrival Delay in Minutes, and Departure Delay in Minutes have much higher values which will affect the model training, hence they need to scaled down. Here due to the high correlations, Departure Delay and Food and drink columns can be dropped as it could lead to overfitting of the model.

```python
from sklearn.preprocessing import MinMaxScaler

scaler1 = MinMaxScaler()
df1["Age"] = scaler1.fit_transform(df1[["Age"]])
df1.head()
```

```python
df1["Flight_Distance"] = scaler1.fit_transform(df1[["Flight_Distance"]])
df1.head()
```

```python
df1["Arrival_Delay_In_Minutes"] = scaler1.fit_transform(df1[["Arrival_Delay_In_Minutes"]])
df1.head()
```

```python
df2 = df1
```

```python
df2.drop(["Departure_Delay_In_Minutes"], inplace=True, axis = 1)
df2.drop(["Food_And_Drink"], inplace=True, axis = 1)
```

# Model Training

```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression()
```

Split the data into training and testing data.

```python
X = df2.drop(["Satisfaction"], axis=1)
Y = df2["Satisfaction"]
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state = 42)
```

```python
model.fit(X_train, Y_train)
```

```python
model = LogisticRegression(max_iter=1000)
model.fit(X_train, Y_train)
```

```python
Y_predicted = model.predict(X_test)
```

Evaluating the model

```python
print(model.score(X_test, Y_test))
```

```python
from sklearn.metrics import confusion_matrix

mat = confusion_matrix(Y_test, Y_predicted)
plt.figure(figsize=(10,7))
sns.heatmap(mat, annot=True)
```

```python
from sklearn.metrics import classification_report

report = classification_report(Y_test, Y_predicted)
print(report)
```
